# Analyzing

```python
# checking the percentage of null values in columns of dataframe-1
pd.set_option('display.max_rows', None)
100*df1.isnull().mean()
```

```python
# dropping the columns from df1 which have null values greater than 50%
for col in df1.columns:
    missing_percent = 100 * df1[col].isnull().mean()
    if missing_percent > 50.0:
        df1.drop(columns=col, inplace=True)
```

```python
# checking the percentage of null values in columns of dataframe-2
pd.set_option('display.max_rows', None)
100*df2.isnull().mean()
```

```python
# dropping the columns from df2 which have null values greater than 30%
for col in df2.columns:
    missing_percent = 100 * df2[col].isnull().mean()
    if missing_percent > 30.0:
        df2.drop(columns=col, inplace=True)
```

```python
# Removing some unnecessary columns from df1
df1 = df1.drop(columns=['NAME_TYPE_SUITE','FLAG_DOCUMENT_2','FLAG_DOCUMENT_3','FLAG_DOCUMENT_4','FLAG_DOCUMENT_5','FLAG_DOCUMENT_6','FLAG_DOCUMENT_7','FLAG_DOCUMENT_8','FLAG_DOCUMENT_9','FLAG_DOCUMENT_10','FLAG_DOCUMENT_11','FLAG_DOCUMENT_12', 'FLAG_DOCUMENT_13', 'FLAG_DOCUMENT_14','FLAG_DOCUMENT_15', 'FLAG_DOCUMENT_16', 'FLAG_DOCUMENT_17','FLAG_DOCUMENT_18', 'FLAG_DOCUMENT_19', 'FLAG_DOCUMENT_20','FLAG_DOCUMENT_21','FLAG_MOBIL','FLAG_EMP_PHONE','FLAG_WORK_PHONE','FLAG_CONT_MOBILE','FLAG_PHONE','FLAG_EMAIL','DAYS_LAST_PHONE_CHANGE','REG_REGION_NOT_LIVE_REGION','REG_REGION_NOT_WORK_REGION','LIVE_REGION_NOT_WORK_REGION','REG_CITY_NOT_LIVE_CITY','REG_CITY_NOT_WORK_CITY','LIVE_CITY_NOT_WORK_CITY','EXT_SOURCE_2','EXT_SOURCE_3','REGION_RATING_CLIENT_W_CITY','YEARS_BEGINEXPLUATATION_AVG','FLOORSMAX_AVG','YEARS_BEGINEXPLUATATION_MODE','FLOORSMAX_MODE','YEARS_BEGINEXPLUATATION_MEDI','FLOORSMAX_MEDI','TOTALAREA_MODE','EMERGENCYSTATE_MODE'])
```

```python
#dropping the unnecessary values from columns of df1
for col in df1.columns:
    if (df1[col] == 'XNA').any() or (df1[col] == 'XAP').any():
        df1[col] = df1[col].replace(['XNA', 'XAP'], np.nan)
```

```python
#dropping the unnecessary values from columns of df2
for col in df2.columns:
    if (df2[col] == 'XNA').any() or (df2[col] == 'XAP').any():
        df2[col] = df2[col].replace(['XNA', 'XAP'], np.nan)
```

# Imputing the missing values

```python
# creating a list of columns (which have numerical values) which are to be imputed and using for loop on them to impute
ls = ['AMT_ANNUITY','AMT_GOODS_PRICE','CNT_FAM_MEMBERS','OBS_30_CNT_SOCIAL_CIRCLE','DEF_30_CNT_SOCIAL_CIRCLE','OBS_60_CNT_SOCIAL_CIRCLE','DEF_60_CNT_SOCIAL_CIRCLE','AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_YEAR','AMT_REQ_CREDIT_BUREAU_QRT']
for col in ls:
    df1[col] = df1[col].fillna(df1[col].median())
```

```python
# creating a list of columns (which have categorical values) which are to be imputed and using for loop on them to impute
ls1 = ['CODE_GENDER','OCCUPATION_TYPE','ORGANIZATION_TYPE']
for col in ls1:
    df1[col] = df1[col].fillna(df1[col].mode()[0])
```

# correcting the column values

```python
ls = ['DAYS_BIRTH','DAYS_EMPLOYED','DAYS_REGISTRATION','DAYS_ID_PUBLISH']
for col in ls:
    df1[col]= df1[col].abs()
```

```python
#creating age column by converting DAYS_BIRTH and DAYS_EMPLOYED to years and dropping them because we have age and years of employed before joining column.
df1['Age'] = df1['DAYS_BIRTH'] // 365
df1.drop(columns='DAYS_BIRTH',inplace=True)
df1['Currently_employed'] = (df1['DAYS_EMPLOYED'] / 365).round(2)
df1.drop(columns='DAYS_EMPLOYED',inplace=True)
```

# Checking Outliers

```python
ls = ['AMT_INCOME_TOTAL','AMT_CREDIT','AMT_ANNUITY','AMT_GOODS_PRICE','Age','Currently_employed','DAYS_ID_PUBLISH']
```

```python
for col in ls:
    sns.boxplot(df1[col])
    plt.title(col)
    plt.show()
```

# What needs to be done to Outliers (if exists)

AMT_INCOME_TOTAL column does not need outliers treatment because income can vary to any extent.

AMT_CREDIT needs outliers treatment.

AMT_ANNUITY needs outliers treatment.

AMT_GOODS_PRICE column does not need outliers treatment because price can vary to any extent.

Age has no outliers.

Currently_employed needs outliers treatment.

DAYS_ID_PUBLISH has no outliers.

```python
ls = ['AMT_CREDIT','AMT_ANNUITY','Currently_employed']
for col in ls:
    # Calculate upper and lower bound for capping and flooring
    upper_bound = df1[col].describe()['75%'] + 1.5 * (df1[col].describe()['75%'] - df1[col].describe()['25%'])
    lower_bound = df1[col].describe()['25%'] - 1.5 * (df1[col].describe()['75%'] - df1[col].describe()['25%'])

    # Cap the upper outliers
    df1.loc[df1[col] > upper_bound, col] = upper_bound

    # Floor the lower outliers
    df1.loc[df1[col] < lower_bound, col] = lower_bound
```

# Univariate Analysis

Univariate Analysis according to Client with payment difficulties and all other cases (Target variable).

```python
# We can divide the dataset into two different dataframes:

# target=1 (clients with payment difficulties)
target1 = df1.loc[df1["TARGET"] == 1]
# target=0 (all other cases)
target0 = df1.loc[df1["TARGET"] == 0]
```

```python
#creating bins for the next step
# creating bins for the age column

bins = [0,30,40,50,60,100]
labels = ['<30','30-40','40-50','50-60','60+']
target1['AGE_RANGE'] = pd.cut(df1.Age, bins=bins, labels=labels)
target0['AGE_RANGE'] = pd.cut(df1.Age, bins=bins, labels=labels)

# Creating bins for income amount

bins = [0,25000,50000,75000,100000,125000,150000,175000,200000,225000,250000,275000,300000,325000,350000,375000,400000,425000,450000,475000,500000,10000000000]
slot = ['0-25000', '25000-50000','50000-75000','75000,100000','100000-125000', '125000-150000', '150000-175000','175000-200000',
       '200000-225000','225000-250000','250000-275000','275000-300000','300000-325000','325000-350000','350000-375000',
       '375000-400000','400000-425000','425000-450000','450000-475000','475000-500000','500000 and above']

target1['AMT_INCOME_RANGE']=pd.cut(target1['AMT_INCOME_TOTAL'],bins,labels=slot)
target0['AMT_INCOME_RANGE']=pd.cut(target0['AMT_INCOME_TOTAL'],bins,labels=slot)

# Creating bins for Credit amount

bins = [0,150000,200000,250000,300000,350000,400000,450000,500000,550000,600000,650000,700000,750000,800000,850000,900000,1000000000]
slots = ['0-150000', '150000-200000','200000-250000', '250000-300000', '300000-350000', '350000-400000','400000-450000',
        '450000-500000','500000-550000','550000-600000','600000-650000','650000-700000','700000-750000','750000-800000',
        '800000-850000','850000-900000','900000 and above']

target1['AMT_CREDIT_RANGE']=pd.cut(target1['AMT_CREDIT'],bins=bins,labels=slots)
target0['AMT_CREDIT_RANGE']=pd.cut(target0['AMT_CREDIT'],bins=bins,labels=slots)
```

```python
categorical_col = ['NAME_CONTRACT_TYPE','CNT_CHILDREN','CODE_GENDER','FLAG_OWN_CAR','FLAG_OWN_REALTY', 'NAME_INCOME_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE','AGE_RANGE','AMT_INCOME_RANGE','AMT_CREDIT_RANGE','ORGANIZATION_TYPE']

for col in categorical_col:
    plt.figure(figsize=(15,6))
    plt.style.use('bmh')
    plt.subplot(1, 2, 1)
    sns.countplot(x = col, data=target0)
    plt.title('Distribution of '+ col +' for target=0', fontsize=13)
    plt.xlabel(col)
    plt.xticks(rotation=90,fontsize=8)
    plt.yticks(fontsize=8)
    plt.ylabel('Number of cases for non-defaulters')
    
    plt.subplot(1, 2, 2)
    sns.countplot(x=col, data=target1)
    plt.title('Distribution of '+ col +' for target=1', fontsize=13)
    plt.xlabel(col)
    plt.xticks(rotation=90,fontsize=8)
    plt.yticks(fontsize=8)
    plt.ylabel('Number of cases for defaulters')
    plt.show()
```
